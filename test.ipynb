# Build a tree-like json structure

```python
from tree_utils import build_nodes 
nodes = build_nodes(df)

nodes[:1]  # quick sanity check
```

```python
import json

print(json.dumps(nodes, indent=2))
```

```python
def print_tree(nodes, indent=0):
    for node in nodes:
        print("  " * indent + f"- {node['label']}")
        if "children" in node:
            print_tree(node["children"], indent + 1)

print_tree(nodes)
```
